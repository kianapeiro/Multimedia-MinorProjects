## Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Image Loading and Visualization

```python
#imread, image: Unequalized_Hawkes_Bay_NZ.jpg
image = cv2.imread('Unequalized_Hawkes_Bay_NZ.jpg', cv2.COLOR_BGR2RGB)
Gimage= cv2.imread('Unequalized_Hawkes_Bay_NZ.jpg', cv2.IMREAD_GRAYSCALE)
#visualization with matplotlib 
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
plt.imshow(Gimage, cmap='gray')
plt.title('Gray Scaled Image')
plt.axis('off')
plt.show()
```

```python
#show the histogram plot of original image
plt.hist(image.ravel(), 256, [0, 256])
plt.title('Histogram of Original Image')
plt.show()
plt.hist(Gimage.ravel(), 256, [0, 256])
plt.title('Histogram of Gray Scaled Image')
plt.show()
```

## Histogram Equalization

```python
#equalize the histogram of image
Eimage = cv2.equalizeHist(Gimage)
#show the histogram plot of equalized image
plt.hist(Eimage.ravel(), 256, [0, 256])
plt.title('Histogram of Global Histogram Equalized Image')
plt.show()
#output visualization
plt.imshow(Eimage, cmap='gray')
plt.title('Global Histogram Equalized Image')
plt.axis('off')
plt.show()
```

## Clahe(adaptive histogram equalization)

first import clache image and analyse its histogram plot

```python
#imread, image: clahe.jpg
clahe_image = cv2.imread('clahe.jpg', cv2.COLOR_BGR2RGB)
clahe_Gimage= cv2.imread('clahe.jpg', cv2.IMREAD_GRAYSCALE)
#visualization with matplotlib 
plt.imshow(clahe_image)
plt.title('Original Image')
plt.axis('off')
plt.show()
plt.imshow(clahe_Gimage, cmap='gray')
plt.title('Gray Scaled Image')
plt.axis('off')
plt.show()
#show the histogram plot of original image
plt.hist(clahe_image.ravel(), 256, [0, 256])
plt.title('Histogram of Original Image')
plt.show()
plt.hist(clahe_image.ravel(), 256, [0, 256])
plt.title('Histogram of Gray Scaled Image')
plt.show()
```

now do the clahe operation and global histogram equalization on clahe's images and analyse the outputs

```python
#global histogram equalization operation
global_equalized = cv2.equalizeHist(clahe_Gimage)
#clahe adaptive histogram equalization operation
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_equalized = clahe.apply(clahe_Gimage)
#output visualization
plt.imshow(global_equalized, cmap='gray')
plt.title('Global Histogram Equalized Image')
plt.axis('off')
plt.show()
plt.imshow(clahe_equalized, cmap='gray')
plt.title('Adaptive Histogram Equalized Image (CLAHE)')
plt.axis('off')
plt.show()
```

```python
#show the histogram plot of global histogram equalization operation result
plt.hist(global_equalized.ravel(), 256, [0, 256])
plt.title('Histogram of Global Histogram Equalized Image')
plt.show()
#show the histogram plot of clahe operation result
plt.hist(clahe_equalized.ravel(), 256, [0, 256])
plt.title('Histogram of Adaptive Histogram Equalized Image (CLAHE)')
plt.show()
```
