## Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Image Loading and Visualization

```python
#imread
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#visualization with matplotlib
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

## Gamma Correction

```python
#implement gamma correction and save it to the new LUT
gamma = 2.2
# Gamma 2.2 has been the standard for Windows and Apple and can produce almost optimal true to life colors with washout or inaccurate shadows.
iGamma = 1.0/gamma
LUT = np.array([((i/255.0)**iGamma)*255 for i in np.arange(0, 256)]).astype("uint8")
```

## LookUPTable mapping

```python
#now map your image with its new LUT which contains correct contrast
cimage = cv2.LUT(image, LUT)
```

## Saving and Visualization

```python
#imsave
cv2.imwrite('gamma corrected image.jpg', cv2.cvtColor(cimage, cv2.COLOR_RGB2BGR))
#visualization with matplotlib
plt.imshow(cimage)
plt.title('Gamma Corrected Image')
plt.axis('off')
plt.show()
```
